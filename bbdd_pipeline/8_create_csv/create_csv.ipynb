```python
group_names = [
  'halflife',
    'cmax',
    'timetocmax',
    'cssmax',
    'auc',
    'clearance',
    'bioavailability',
    'maximaltolerateddose'

]

```

```python
import os
import sqlite3
import pandas as pd

# Conexión a la base de datos DRUG_DB_applied.db
conn_drug_db = sqlite3.connect("DRUG_DB.db")

# Consulta para obtener los nombres de las tablas (nombres de las drogas) en DRUG_DB_applied.db
query_tables = "SELECT name FROM sqlite_master WHERE type='table';"
tables = conn_drug_db.execute(query_tables).fetchall()
for table in tables:
    print(table[0])
# Conexión a la base de datos DB_DESCRIPTORS.db
conn_descriptors = sqlite3.connect("BD_DESCRIPTORS.db")

# Nombres de los grupos de parámetros


# Iterar sobre los nombres de los grupos
for group_name in group_names:
   
    parameter_csv_data = {"name": [], "mean": [], "std": [], "median": []}
    tables_added=[]
    for table in tables:
        
        if table[0] not in tables_added:
            tables_added.append(table[0])
            # Consulta para obtener los parámetros, la media, la desviación estándar y la mediana desde DRUG_DB_applied.db
            query_params = f"""SELECT drugbank_id, parameter, mean, std, median FROM "{table[0]}" WHERE parameter = '{group_name}';"""
            parameters = conn_drug_db.execute(query_params).fetchall()

            # Construir el diccionario para el .csv para este grupo
            
            if parameters:    
                drugbank_id = parameters[0][0]

                print(drugbank_id)
                # Consultar los descriptores correspondientes desde DB_DESCRIPTORS.db usando drugbank_id
                query_descriptors = f"""
                    SELECT descriptor_name, descriptor_value 
                    FROM "{drugbank_id}"
                    WHERE descriptor_name IN (SELECT DISTINCT descriptor_name FROM "{drugbank_id}");
                """
                try:
                    descriptors = conn_descriptors.execute(query_descriptors).fetchall()
                except:
                    pass
                else:
                
                    # Agregar los descriptores al diccionario
                    for descriptor in descriptors:
                        descriptor_name = descriptor[0]
                        descriptor_value = descriptor[1]
                        #if table[0] not in parameter_csv_data['name']:
                        
                        parameter_csv_data.setdefault(descriptor_name, []).append(descriptor_value)
                        
                    
                    parameter_name = parameters[0][1]
                    parameter_csv_data["name"].append(table[0])
                    parameter_csv_data["mean"].append(parameters[0][2])
                    parameter_csv_data["std"].append(parameters[0][3])
                    parameter_csv_data["median"].append(parameters[0][4])
                    
                    
                
                
            
                
    # Crear un DataFrame a partir del diccionario y guardar como CSV
    parameter_df = pd.DataFrame(parameter_csv_data)
    parameter_csv_filename = f"{group_name}.csv"
    parameter_csv_filename = os.path.join("csv", parameter_csv_filename)
    parameter_df.to_csv(parameter_csv_filename, index=False)

# Cerrar las conexiones a las bases de datos
conn_drug_db.close()
conn_descriptors.close()
```
